```python
Analysing player 15 with the event data
```

```python
import pandas as pd
```

```python
player_data = pd.read_csv('../sample-data/data/Sample_Game_1/player_15_data.csv')
event_data = pd.read_csv('../sample-data/data/Sample_Game_1/Sample_Game_1_RawEventsData.csv')
```

```python
player = "Player15"

player_events = event_data[
    (event_data["From"] == player) |
    (event_data["To"] == player)
].copy()
player_events.reset_index(drop=True, inplace=True)
player_events = player_events.iloc[:, 1:]  
player_events.head()
```

```python
player_events.info()
```

```python
event_like = player_events[[
    'Type',
    'Subtype', 
    'Start Frame',
    'Start X', 
    'Start Y'
    ]].rename(columns={
        'Type': 'event_type',
        'Subtype': 'event_subtype',
        'Start Frame': 'frame',
        'Start X': 'x',
        'Start Y': 'y'
    })
event_like.head()
```

```python
player_data.head()
```

```python
player_data = player_data[[
    'Frame',
    'player15_x', 
    'player15_y'
    ]].rename(columns={
        'Frame': 'frame',
        'player15_x': 'x',
        'player15_y': 'y'
    })
player_data.head()
```

```python
merged_data = player_data.merge(
    event_like,
    on=['frame'],
    how='left',
    suffixes=('_player', '_event')
)
merged_data.head(20)
```

```python
merged_event_data = player_data.merge(
    event_like,
    on=['frame'],
    how='right',
    suffixes=('_player', '_event')
)
merged_event_data.head(20)
```

```python
# sanity check
dx_m = (merged_event_data["x_player"] - merged_event_data["x_event"]) * 105
dy_m = (merged_event_data["y_player"] - merged_event_data["y_event"]) * 68
merged_event_data['event_tracking_dist'] = (
    (dx_m) ** 2 +
    (dy_m) ** 2
).apply(lambda x: x ** 0.5)
merged_event_data[['frame', 'event_type', 'event_tracking_dist']].head(20)
```
